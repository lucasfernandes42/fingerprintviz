```python
extractor = list()
img_size = list()
resolution = list()
img_pixeis = list()
```

```python
for img in data["input"]:
    if img.find("DB1") > -1:
        extractor.append("optical sensor CrossMatch")
        img_size.append("640x480")
        resolution.append(500)
    if img.find("DB2") > -1:
        extractor.append("optical sensor Digital Persona")
        img_size.append("328x364")
        resolution.append(500)
    if img.find("DB3") > -1:
        extractor.append("thermal sweeping sensor Atmel")
        img_size.append("300x480")
        resolution.append(512)
    if img.find("DB4") > -1:
        extractor.append("synthetic fingerprint generation")
        img_size.append("288x384")
        resolution.append(500)
```

```python
resolution
```

```python
data["resolution_(DPI)"] = resolution
data["image_size"] = img_size
data["sensor_type"] = extractor
```
