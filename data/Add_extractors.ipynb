```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("fingerprints_quality_fvc_2004.csv")
data
```

```python
extractor = list()
img_size = list()
resolution = list()
img_pixeis = list()
```

```python
for img in data["input"]:
    if img.find("DB1") > -1:
        extractor.append("optical sensor V300 by CrossMatch")
        img_size.append("640x480")
        resolution.append(500)
    if img.find("DB2") > -1:
        extractor.append("optical sensor U.are.U 4000 by Digital Persona")
        img_size.append("328x364")
        resolution.append(500)
    if img.find("DB3") > -1:
        extractor.append("thermal sweeping sensor FingerChip FCD4B14CB by Atmel")
        img_size.append("300x480")
        resolution.append(512)
    if img.find("DB4") > -1:
        extractor.append("synthetic fingerprint generation SFinGe v3.0")
        img_size.append("288x384")
        resolution.append(500)
```

```python
resolution
```

```python
data["resolution_(DPI)"] = resolution
data["image_size"] = img_size
data["sensor_type"] = extractor
```
